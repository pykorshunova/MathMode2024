4. Создайте массив нулей размера 10 x 10, а затем создайте в нём “рамку” из единиц по краям.

```python
ar = np.array([[0.]*10]*10)

for i in range(len(ar)):
    for j in range(len(ar[i])):
        if i == 0 or i == 9:
            ar[i][j] = 1
        elif j == 0 or j ==9:
            ar[i][j] = 1
            
print(ar)
```

5. Создайте массив 8 x 8 с шахматной доской из нулей и единиц.  

```python
ar = np.array([[0.]*8]*8)
        
for i in range(len(ar)):
    for j in range(len(ar[i])):
        if i % 2 == 0 and j % 2 == 1:
            ar[i][j] = 1
        if i % 2 == 1 and j % 2 == 0:
            ar[i][j] = 1

print(ar)
```

6. Создайте функцию, которая создает $n×n$ матрицу с $(i,j)$-элементом, равным $i+j$.



```python
def create_array(n):
    ar = np.array([[0.]*n]*n)
    for i in range(len(ar)):
        for j in range(len(ar[i])):
            ar[i][j] = i + j
    return ar
            
print(create_array(5))
```

1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 250)
y = x**2 * np.sin(1/x**2) + x

plt.plot(x, y, label= "$x^2⋅sin(1/x^2)+x$")
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции $x^2⋅sin(1/x^2)+x$')
plt.legend()
plt.grid(True)
plt.show()
```

2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)

```python
x = np.linspace(5, 25, 200)
y_1 = 1/(1 + x**2)
y_2 = 1/x**2
relative_diff = abs(1 - (y_1/y_2))

plt.title('Относительная разница функций $1/(1+x^2)$ и $1/x^2$')
plt.semilogy(x, relative_diff)
plt.xlabel('Линейная шкала для оси x')
plt.ylabel('Логорифмическая шкала для оси y')
plt.grid()
plt.show()
```

4. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)

plt.figure()
plt.imshow(Z, extent=[-np.pi, np.pi, -np.pi, np.pi], cmap='magma')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции sin(x)sin(y)')
plt.show()
```

4. Посчитайте две матрицы: $B = A^TA$ и $C = AA^T$. Что вы можете о них сказать? Какого они размера? В чём их сходство и отличие? Найдите сумму диагональных элементов матриц $B$ и $C$. Посчитайте детерминант и ранг матриц $A$, $B$ и $C$, посмотрите на спектр (набор собственных значений) матриц $B$ и $C$, какие выводы вы можете сделать? (подсказка: используйте готовые функции из библиотеки `np.linalg`). 

```python
A = np.random.rand(3, 3)
A_t = A.transpose()
B = np.dot(A_t, A)
C = np.dot(A, A_t)
print(A)
print(B)
```

```python
print("След матрицы B: ", np.trace(B))
print("След матрицы C: ", np.trace(C))
```

```python
print("Детерминант матрицы A: ", np.linalg.det(A))
print("Детерминант матрицы B: ", np.linalg.det(B))
print("Детерминант матрицы C: ", np.linalg.det(C))
```

```python
print("Ранг матрицы A: ", np.linalg.matrix_rank(A))
print("Ранг матрицы B: ", np.linalg.matrix_rank(B))
print("Ранг матрицы C: ", np.linalg.matrix_rank(C))
```

eigenvalues - это собственные значения, eigenvectors - это собственные векторы

```python
np.linalg.eig(A)
```

```python
print("Спектр матрицы A: ", np.linalg.eig(A)[0])
print("Спектр матрицы B: ", np.linalg.eig(B)[0])
print("Спектр матрицы C: ", np.linalg.eig(C)[0])
```

Вывод: у матриц B и C одинаковые размеры, суммы элементов главной диагонали, детерминант, ранг и наборы собственных значений
